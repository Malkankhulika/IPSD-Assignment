```python
movie = film.copy(deep=True)
```

```python
movie.dropna(inplace=True)
```

```python
movie
```

```python
movie.isna().sum()
```

4. Drop missing value berdasarkan baris, kolom, imputasi dengan mean, modus, median

a. berdasarkan baris

```python
baris = movie.copy(deep=True)
drop_baris = baris.dropna(axis=0)
```

b. berdasarkan kolom

```python
kolom = movie.copy(deep=True)
drop_kolom = kolom.dropna(axis=1)
```

c. berdasarkan imputasi mean

```python
mean_imputation = movie.copy(deep=True)
mean_imputation["Time_taken"] = mean_imputation["Time_taken"].fillna(mean_imputation["Time_taken"].mean())
```

```python
mean_imputation.isna().sum()
```

d. berdasarkan Imputasi dengan Modus

```python
modus_imputation = movie.copy(deep=True)
modus_imputation["Time_taken"] = modus_imputation["Time_taken"].fillna(modus_imputation["Time_taken"].mode()[0])
```

```python
median_imputation = movie.copy(deep=True)

median_imputation["Time_taken"] = median_imputation["Time_taken"].fillna(median_imputation["Time_taken"].median())
```
